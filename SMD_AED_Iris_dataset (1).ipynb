```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Carregando os dados do Iris dataset
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                  columns= iris['feature_names'] + ['target'])

# 1. Histogramas e Gráficos de Distribuição
# Os histogramas mostram a distribuição de cada variável.
# Podemos observar a distribuição de valores e identificar possíveis outliers.
df.hist()
plt.show()
print("Histogramas e Gráficos de Distribuição: Mostram a distribuição de cada variável.")

# 2. Gráficos de Dispersão (Scatter Plots)
# Os scatter plots mostram a relação entre pares de variáveis.
# Eles nos ajudam a visualizar se há correlação entre as variáveis e se há agrupamentos distintos de pontos.
sns.pairplot(df, diag_kind='kde', hue='target')
plt.show()
print("Gráficos de Dispersão (Scatter Plots): Mostram a relação entre pares de variáveis.")

# 3. Matriz de Correlação
# A matriz de correlação mostra a relação entre todas as variáveis do dataset.
# Valores próximos de 1 indicam uma forte correlação positiva, valores próximos de -1 indicam uma forte correlação negativa e valores próximos de 0 indicam pouca ou nenhuma correlação.
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
print("Matriz de Correlação: Mostra a relação entre todas as variáveis do dataset.")

# 4. Box Plots
# Os box plots mostram a distribuição de cada variável, permitindo identificar outliers e entender a distribuição dos dados.
sns.boxplot(data=df, orient="h")
plt.show()
print("Box Plots: Mostram a distribuição de cada variável.")

# 5. Gráficos de Barras e Gráficos de Pizza
# Os gráficos de barras mostram a contagem de cada classe no dataset.
# Eles nos ajudam a entender a distribuição das classes e se o dataset está balanceado.
sns.countplot(data=df, x='target')
plt.show()
print("Gráficos de Barras e Gráficos de Pizza: Mostram a contagem de cada classe no dataset.")

# 6. Heatmap
# O heatmap é uma representação visual da matriz de correlação.
# Ele nos ajuda a identificar padrões de correlação entre as variáveis.
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
print("Heatmap: Representa visualmente a matriz de correlação.")

# 7. Pair Plots
# Os pair plots mostram scatter plots para todas as combinações de variáveis numéricas e histogramas para as variáveis individuais.
# Eles nos ajudam a entender a relação entre as variáveis e identificar possíveis padrões nos dados.
sns.pairplot(df, kind='scatter')
plt.show()
print("Pair Plots: Mostram scatter plots para todas as combinações de variáveis numéricas.")

# 8. Análise de Componentes Principais (PCA) - Este exemplo é apenas ilustrativo, não é uma visualização direta
# A análise de componentes principais (PCA) é uma técnica de redução de dimensionalidade que nos permite visualizar a estrutura dos dados em um espaço de dimensão reduzida.
# Este gráfico de dispersão mostra os dados projetados nos dois primeiros componentes principais.
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(df.drop('target', axis=1))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=df['target'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA')
plt.show()
print("Análise de Componentes Principais (PCA): Mostra os dados projetados nos dois primeiros componentes principais.")
```
